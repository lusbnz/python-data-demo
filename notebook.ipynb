```python
import pandas as pd

competition_path = 'raw_data/competition.json'
club_path = 'raw_data/club.json'
player_path = 'raw_data/player.json'

competition_data = pd.read_json(competition_path)
club_data = pd.read_json(club_path)
player_data = pd.read_json(player_path)
```

```python
competition_data.info()
club_data.info()
player_data.info()
```

```python
clubs_list = []
for index, row in competition_data.iterrows():
    for club in row['clubs']:
        clubs_list.append({
            'club_id': club['id'],
            'club_name': club['name'],
            'competition_name': row['name'],
            'competition_id': row['id'],
            'season_id': row['seasonID'],
        })

competition_df = pd.DataFrame(clubs_list)

print("Competition DataFrame:")
print(competition_df)
competition_df.to_csv('competition_df.csv', index=False)
```

```python
club_list = []
player_list = []

for club_id, club_info in club_data.items():
    club_profile = club_info['profile']
    league_info = club_profile['league']
    squad_info = club_profile['squad']
    club_list.append({
        'club_id': club_profile['id'],
        'club_name': club_profile['name'],
        'stadium_name': club_profile['stadiumName'],
        'stadiumSeats': club_profile['stadiumSeats'],
        'average_age': squad_info['averageAge'],
        'name': league_info['name'],
        'competition_id': league_info['id'],
    })
    
    for player in club_info['players']['players']:
        player_list.append({
            'player_id': player['id'],
            'player_name': player['name'],
            'position': player['position'],
            'age': player['age'],
            'nationality': ', '.join(player['nationality']),
            'height': player.get('height', 'N/A'),
            'foot': player.get('foot', 'N/A'),
            'contract': player.get('contract', 'N/A'),
            'market_value': player.get('marketValue', 'N/A'),
            'club_id': club_profile['id']
        })

clubs_df = pd.DataFrame(club_list)
players_df = pd.DataFrame(player_list)

print("Clubs DataFrame:")
print(clubs_df)
print("\nPlayers DataFrame:")
print(players_df)

clubs_df.to_csv('club_profile.csv', index=False)
players_df.to_csv('club_player.csv', index=False)
```

```python
player_list = []

for player_id, player_info in player_data.items():
    profile = player_info.get('profile', {})
    market_value = player_info.get('market_value', {})
    transfers = player_info.get('transfers', {}).get('transfers', [])
    stats = player_info.get('stats', {}).get('stats', [])
    injuries = player_info.get('injuries', {}).get('injuries', [])

    player_details = {
        'player_id': profile.get('id', 'N/A'),
        'player_name': profile.get('name', 'N/A'),
        'date_of_birth': profile.get('dateOfBirth', 'N/A'),
        'age': profile.get('age', 'N/A'),
        'height': profile.get('height', 'N/A'),
        'citizenship': ', '.join(profile.get('citizenship', [])),
        'foot': profile.get('foot', 'N/A'),
        'position': profile.get('position', {}).get('main', 'N/A'),
        'shirt_number': profile.get('shirtNumber', 'N/A'),
        'club_name': profile.get('club', {}).get('name', 'N/A'),
        'club_id': profile.get('club', {}).get('id', 'N/A'),
        'joined_on': profile.get('club', {}).get('joined', 'N/A'),
        'contract_expires': profile.get('club', {}).get('contractExpires', 'N/A'),
        'market_value': market_value.get('marketValue', 'N/A'),
        'injury_history': []
    }

    player_summary = {
        'player_id': profile.get('id', 'N/A'),
        'player_name': profile.get('name', 'N/A'),
        'position': profile.get('position', {}).get('main', 'N/A'),
        'age': profile.get('age', 'N/A'),
        'height': profile.get('height', 'N/A'),
        'foot': profile.get('foot', 'N/A'),
        'market_value': market_value.get('marketValue', 'N/A'),
        'contract_expires': profile.get('club', {}).get('contractExpires', 'N/A'),
        'club_id': profile.get('club', {}).get('id', 'N/A'),
        'joined_on': profile.get('club', {}).get('joined', 'N/A'),
    }

    for history in market_value.get('marketValueHistory', []):
        age = history.get("age", "N/A")
        value = history.get('value', "N/A")
        date = history.get('date', "N/A")

        player_details.update({
            f'market_value_at_age_{age}': value,
            f'market_value_date_{age}': date,
        })

    for transfer in transfers:
        player_summary.update({
            'transfer_from': transfer.get('from', {}).get('clubName', 'N/A'),
            'transfer_to': transfer.get('to', {}).get('clubName', 'N/A'),
            'transfer_date': transfer.get('date', 'N/A'),
            'transfer_market_value': transfer.get('marketValue', 'N/A'),
        })

    for stat in stats:
        competition_name = stat.get("competitionName", "Unknown Competition")
        player_summary.update({
            f'{competition_name}_appearances': stat.get('appearances', 'N/A'),
            f'{competition_name}_minutes_played': stat.get('minutesPlayed', 'N/A'),
        })

    if injuries:
        injury_details = []
        for injury in injuries:
            injury_details.append({
                'injury_date': injury.get('date', 'N/A'),
                'injury_description': injury.get('description', 'N/A'),
                'injury_duration': injury.get('duration', 'N/A'),
            })
        player_details['injury_history'] = injury_details

    player_list.append(player_summary)
    player_list[-1].update(player_details)

players_df = pd.DataFrame(player_list)

print("Players DataFrame:")
print(players_df)

players_df.to_csv('players_df.csv', index=False)
```
